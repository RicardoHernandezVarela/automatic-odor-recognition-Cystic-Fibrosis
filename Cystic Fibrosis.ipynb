```python
CF = CF.drop(['Infectionstatus'], axis=1).reset_index(drop=True)
kontrol = kontrol.drop(['Infectionstatus'], axis=1).reset_index(drop=True)
```

```python
def split_data(df, n):
    sample = df.sample(n).index
    train = df.iloc[sample]
    test = df.iloc[[i for i in df.index if i not in sample]]
    
    return train, test
```

```python
def class_atr_split(df1, df2, target):
    conc_set = pd.concat([df1, df2])
    classes = conc_set[target] 
    atrb = conc_set.drop([target], axis=1)
    
    return atrb, classes
```

```python
def validation(df1, df2, n1, n2, target, iterations, clf):
    clf_name = clf.__class__.__name__
    #print(clf_name)
    
    average_results = [[], [], []]
    
    for i in range(0, iterations):
        results = []
        
        train_df1, test_df1 = split_data(df1, n1)
        train_df2, test_df2 = split_data(df2, n2)

        x_train, y_train = class_atr_split(train_df1, train_df2, target)
        x_test, y_test = class_atr_split(test_df1, test_df2, target)
        
        clf.fit(x_train, y_train)
        score = clf.score(x_test, y_test)
        average_results[0].append(score)
        y_pred = clf.predict(x_test)
        
        conf_matrix = confusion_matrix(y_test, y_pred)
        
        sensitivity = conf_matrix[0][0]/(conf_matrix[0][0]+conf_matrix[0][1])
        average_results[1].append(sensitivity)
        
        specificity = conf_matrix[1][1]/(conf_matrix[1][0]+conf_matrix[1][1])
        average_results[2].append(specificity)
        
        results.extend((score, sensitivity, specificity))
        #print(results)
        #print(conf_matrix) 
    
    average_results = pd.DataFrame(data = average_results)
    average_results['avg'] = average_results.mean(axis=1)
    average_results['std'] = average_results.std(axis=1)
    return average_results
```

```python
from sklearn.metrics import confusion_matrix, classification_report, make_scorer, accuracy_score
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC, LinearSVC, NuSVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier, GradientBoostingClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis
```

```python
classifiers = [
    KNeighborsClassifier(3),
    SVC(kernel="rbf", C=0.025, probability=True, gamma='auto'),
    DecisionTreeClassifier(),
    RandomForestClassifier(),
    AdaBoostClassifier(),
    GradientBoostingClassifier(),
    GaussianNB(),
    LinearDiscriminantAnalysis()]
```

```python
clf_avg = validation(CF, kontrol, 18, 13, 'Group', 10, classifiers[2])
clf_avg
```

```python
results = []
for clf in classifiers:
    clf_avg = validation(CF, kontrol, 39, 13, 'Group', 10, clf)
    performance = pd.concat([clf_avg['avg'], clf_avg['std']]).reset_index(drop=True)
    #print(performance)
    results.append(performance)
```

```python
results = pd.DataFrame(data = results)
results
```

```python
results.columns = ["precisión", "sensitividad", "especificidad", " std precisión", " std sensitividad", "std especificidad"]
clf_names = [clf.__class__.__name__ for clf in classifiers]
results['Clasificador'] = clf_names
results
```

```python
results = results[['Clasificador',"precisión", "sensitividad", "especificidad", " std precisión", " std sensitividad", "std especificidad"]]
results
```

```python
rest = []
for i in range(0, 15):
    results = []
    
    for clf in classifiers:
        clf_avg = validation(CF, kontrol, 39, 13, 'Group', 10, clf)
        performance = pd.concat([clf_avg['avg'], clf_avg['std']]).reset_index(drop=True)
        #print(performance)
        results.append(performance)
        
    results = pd.DataFrame(data = results)
    rest.append(results)
```
