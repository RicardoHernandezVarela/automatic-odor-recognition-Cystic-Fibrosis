```python
import pandas as pd
import sklearn
import numpy as np
import matplotlib.pyplot as plt

#cargar el dataset en un DataFrame de Pandas
data = pd.read_csv('enose_data.csv',delimiter=",")
```

```python
print(data['Group'].count())
print(data['Group'].value_counts())
```

```python
print(data['Infectionstatus'].count())
print(data['Infectionstatus'].value_counts())
```

```python
grupos = data['Group'].values
print(grupos[0])

infstatus = data['Infectionstatus'].values
print(infstatus[0])
```

```python
sensor_data = data.iloc[:, 0:-4]
print(sensor_data)
```

```python
grupos_data = pd.concat([sensor_data, data[['Group']]], axis = 1)
print(grupos_data.head())
```

```python
grupos_data.to_csv('grupos_data.csv', index=False)
```

```python
infstatus_data = pd.concat([sensor_data, data[['Infectionstatus']]], axis = 1)

infstatus_data = [infstatus_data[['Infectionstatus']] != 'BC']
print(infstatus_data.head())
print(infstatus_data['Infectionstatus'].value_counts())
```

```python
infstatus_data.to_csv('infstatus_data.csv', index=False)
```
